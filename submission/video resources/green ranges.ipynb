```python
import cv2
from math import pow, atan2, sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
    cv_image=cv2.imread('./GreenDetectorImages/mailbox.png')
    # convert to hsv
    hsv = cv2.cvtColor(cv_image, cv2.COLOR_BGR2HSV)
    # get the green of the utility bot
    green = np.uint8([[[5, 5, 255]]])
    hsvGreen = cv2.cvtColor(green, cv2.COLOR_BGR2HSV)
    # create a mask to isolate the green in the cart
    mask_g = cv2.inRange(hsv,(60, 255, 100), (100, 255,255))

    if (np.all((mask_g == 0))):
        print('no')
    else:
        print("Yes")
        
    plt.axis("off")    
    plt.imshow(cv2.cvtColor(cv_image, cv2.COLOR_BGR2RGB))
    plt.show()
```
